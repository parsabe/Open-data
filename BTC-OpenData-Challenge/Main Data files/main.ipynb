```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json
```

```python
data = pd.read_csv('BTC-USD.csv')
df = pd.DataFrame(data)
df.head()
```

```python
date = df["Date"]
date = pd.to_datetime(date)
df["year"] = date.dt.year
df["month"] = date.dt.month
df["day"] = date.dt.day
df['new_mon'] = date.dt.month_name()
```

```python
df.head()
```

```python
df.dropna().copy()
```

```python
df.drop(columns=['Date'], inplace=True)
```

```python
df.rename(columns={'new_mon': 'month_name'}, inplace=True)
df = df[df['year'] != 2014]
```

```python
df.to_csv("btc_clean.csv", index=False)

# (Optional) Save also as JSON
df.to_json("btc_clean.json", orient="records", date_format="iso", indent=2)
```
